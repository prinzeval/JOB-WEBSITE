```python
from time import sleep
from selenium.webdriver.common.by import By
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

# Prompt user for job title and location
job_title = input("Enter the job title you're looking for (e.g., 'data analyst entry level'): ")
location = input("Enter the job location (e.g., 'Remote' or 'New York'): ")

# Format the inputs for URL query
formatted_job_title = job_title.replace(' ', '+')
formatted_location = location.replace(' ', '+')

# Construct Indeed search URL based on user input
main_link = f'https://www.indeed.com/jobs?q={formatted_job_title}&l={formatted_location}'

# Initialize WebDriver
driver = webdriver.Chrome()  # Make sure ChromeDriver is properly set up
driver.get(main_link)
sleep(2)

# Find job cards
indeed_cards = driver.find_elements(By.XPATH, '//div[@class="job_seen_beacon"]')
print(indeed_cards)

# Initialize lists for storing job details
my_card_link_list = []
heads = "https://www.indeed.com"

# Extract job links
for WebElement in indeed_cards:
    sleep(3)
    elementHTML = WebElement.get_attribute('outerHTML')
    elementsoup = BeautifulSoup(elementHTML, 'html5lib')
    card_link = elementsoup.find('a', href=True)
    if card_link:
        card_join = card_link['href']
        my_card_link = heads + card_join
        my_card_link_list.append(my_card_link)

print("Job Links:", my_card_link_list)

# Close the driver
driver.quit()
```

```python
import csv
import pymysql

# MySQL database connection details
HOST = 'localhost'
USER = 'root'
PASSWORD = 'Vondabaic2020'
DATABASE = 'val_data'

# Function to establish MySQL connection
def connect_to_database():
    try:
        connection = pymysql.connect(
            host=HOST,
            user=USER,
            password=PASSWORD,
            database=DATABASE,
            cursorclass=pymysql.cursors.DictCursor
        )
        return connection
    except pymysql.Error as e:
        print(f"Error connecting to MySQL database: {e}")
        return None

# Function to read CSV file and insert data into database
def insert_data_from_csv(csv_file):
    try:
        # Establish connection to MySQL database
        connection = connect_to_database()
        if connection is None:
            return

        # Open CSV file and create CSV reader object
        with open(csv_file, 'r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            
            # Iterate over each row in the CSV file
            for row in reader:
                # Truncate job_link value if it exceeds maximum length
                job_link = row['JOB LINK'][:255]  # Assuming the maximum length is 255 characters

                # Prepare INSERT statement
                insert_query = """
                    INSERT INTO job_postings (position, company_name, location, salary, job_link, benefits, description, employment_type)
                    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
                """
                values = (
                    row['POSITION'],
                    row['COMPANY NAME'],
                    row['LOCATION'],
                    row['SALARY'],
                    job_link,  # Use truncated job_link value
                    row['BENEFITS'],
                    row['DESCRIPTION'],
                    row['EMPLOYMENT TYPE']
                )

                # Execute INSERT statement
                with connection.cursor() as cursor:
                    cursor.execute(insert_query, values)

        # Commit changes to the database
        connection.commit()
        print("Data inserted successfully!")

    except Exception as e:
        print(f"Error inserting data into database: {e}")

    finally:
        # Close database connection
        if connection:
            connection.close()

# Path to your CSV file
csv_file_path = r'C:\Users\valen\Desktop\JOB-WEBSITE\output.csv'

# Call function to insert data from CSV into database
insert_data_from_csv(csv_file_path)
```
