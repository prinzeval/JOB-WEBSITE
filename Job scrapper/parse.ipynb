```python
from time import sleep
from selenium.webdriver.common.by import By
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

# Prompt user for job title and location
job_title = input("Enter the job title you're looking for (e.g., 'data analyst entry level'): ")
location = input("Enter the job location (e.g., 'Remote' or 'New York'): ")

# Format the inputs for URL query
formatted_job_title = job_title.replace(' ', '+')
formatted_location = location.replace(' ', '+')

# Construct Indeed search URL based on user input
main_link = f'https://www.indeed.com/jobs?q={formatted_job_title}&l={formatted_location}'

# Initialize WebDriver
driver = webdriver.Chrome()  # Make sure ChromeDriver is properly set up
driver.get(main_link)
sleep(2)

# Find job cards
indeed_cards = driver.find_elements(By.XPATH, '//div[@class="job_seen_beacon"]')
print(indeed_cards)

# Initialize lists for storing job details
my_card_link_list = []
heads = "https://www.indeed.com"

# Extract job links
for WebElement in indeed_cards:
    sleep(3)
    elementHTML = WebElement.get_attribute('outerHTML')
    elementsoup = BeautifulSoup(elementHTML, 'html5lib')
    card_link = elementsoup.find('a', href=True)
    if card_link:
        card_join = card_link['href']
        my_card_link = heads + card_join
        my_card_link_list.append(my_card_link)

print("Job Links:", my_card_link_list)

# Close the driver
driver.quit()

from time import sleep
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome() 

# Lists to store extracted data
position_list = []
company_name_list = []
location_list = []
salary_list = []
var_job_link_list = []
benefit_list = []
description_list = []
employment_type_list = []

# Loop through each job link
for lines in my_card_link_list:
    driver.get(lines)
    sleep(2)
    html_content = driver.page_source
    sharksoup = BeautifulSoup(html_content, 'html.parser')
    
    # Position
    linkz = sharksoup.find('div', {'class': 'jobsearch-InfoHeaderContainer jobsearch-DesktopStickyContainer css-zt53js eu4oa1w0'})
    position = linkz.find('h1').text.strip() if linkz and linkz.find('h1') else "NULL"
    position_list.append(position)

    # Company Name
    company_name_element = linkz.find('span', {'class': "css-1saizt3 e1wnkr790"})
    company_name = company_name_element.text.strip() if company_name_element else "NULL"
    company_name_list.append(company_name)

    # Location
    location = linkz.find('div', {'class': "css-waniwe eu4oa1w0"}).text.strip() if linkz and linkz.find('div', {'class': "css-waniwe eu4oa1w0"}) else "NULL"
    location_list.append(location) 

    # Salary
    salary = linkz.find('span', {'class': "css-19j1a75 eu4oa1w0"})
    salary_text = salary.text.strip() if salary else "NULL"
    salary_list.append(salary_text)

    # Job Link (Updated logic to get the link)
    apply_link_container = sharksoup.find('div', {'id': 'viewJobButtonLinkContainer', 'class': 'icl-u-lg-inlineBlock viewJobButtonLinkContainer css-aunbg2 eu4oa1w0'})
    job_link = None

    if apply_link_container:
        # Find the button tag within the container
        button_tag = apply_link_container.find('button', class_='css-1oxck4n e8ju0x51')
        # Extract the 'href' attribute (custom attribute in this case)
        job_link = button_tag.get('href') if button_tag else "No job link found"
        var_job_link_list.append(job_link)

    # Employment Type
    employment = sharksoup.find('div', {'class': 'css-1xkrvql eu4oa1w0'})
    employ = employment.find('span', {'class': "css-k5flys eu4oa1w0"}).text.strip() if employment and employment.find('span', {'class': "css-k5flys eu4oa1w0"}) else "NULL"
    employment_type_list.append(employ)

    # Benefits
    benefit = sharksoup.find('div', {'id': "benefits", 'data-testid': "benefits-test", 'class': "css-jaxxzt eu4oa1w0"})
    benefits_list = []
    if benefit:
        ul_element = benefit.find('ul', {'class': 'css-8tnble eu4oa1w0'})
        if ul_element:
            benefits_list = [li_element.text.strip() for li_element in ul_element.find_all('li')]
    benefit_list.append(benefits_list if benefits_list else "No benefits found.")

    # Description
    description_div = sharksoup.find('div', {'id': "jobDescriptionText", 'class': "jobsearch-JobComponent-description css-16y4thd eu4oa1w0"})
    description = description_div.text.strip() if description_div else "NULL"
    description_list.append(description)

Job_link_list = []

# Configure ChromeDriver
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.199 Safari/537.36")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Process job links
for job_link in var_job_link_list:
    try:
        driver.get(job_link)

        # Wait for the redirection to complete
        WebDriverWait(driver, 20).until(
            EC.url_changes(job_link)
        )
        time.sleep(1)  # Additional buffer time

        # Capture redirected URL
        redirected_url = driver.current_url
        Job_link_list.append(redirected_url)
        
    except Exception as e:
        print(f"Error processing {job_link}: {e}")

# Close the browser
driver.quit()

# Output lists
print("Position List:", position_list)
print("Company Name List:", company_name_list)
print("Location List:", location_list)
print("Salary List:", salary_list)
print("Job Link List:", Job_link_list)
print("Benefit List:", benefit_list)
print("Description List:", description_list)
print("Employment Type List:", employment_type_list)

TABLE = {
    'POSITION': position_list,
    'COMPANY NAME': company_name_list,
    "LOCATION": location_list,
    'SALARY': salary_list,
    "JOB LINK": Job_link_list,
    "BENEFITS": benefit_list,
    'DESCRIPTION': description_list,
    'EMPLOYMENT TYPE': employment_type_list
}
daf = pd.DataFrame(TABLE)
print(daf)
```

```python
import csv
import pymysql

# MySQL database connection details
HOST = 'localhost'
USER = 'root'
PASSWORD = 'Vondabaic2020'
DATABASE = 'val_data'

# Function to establish MySQL connection
def connect_to_database():
    try:
        connection = pymysql.connect(
            host=HOST,
            user=USER,
            password=PASSWORD,
            database=DATABASE,
            cursorclass=pymysql.cursors.DictCursor
        )
        return connection
    except pymysql.Error as e:
        print(f"Error connecting to MySQL database: {e}")
        return None

# Function to read CSV file and insert data into database
def insert_data_from_csv(csv_file):
    try:
        # Establish connection to MySQL database
        connection = connect_to_database()
        if connection is None:
            return

        # Open CSV file and create CSV reader object
        with open(csv_file, 'r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            
            # Iterate over each row in the CSV file
            for row in reader:
                # Truncate job_link value if it exceeds maximum length
                job_link = row['JOB LINK'][:255]  # Assuming the maximum length is 255 characters

                # Prepare INSERT statement
                insert_query = """
                    INSERT INTO job_postings (position, company_name, location, salary, job_link, benefits, description, employment_type)
                    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
                """
                values = (
                    row['POSITION'],
                    row['COMPANY NAME'],
                    row['LOCATION'],
                    row['SALARY'],
                    job_link,  # Use truncated job_link value
                    row['BENEFITS'],
                    row['DESCRIPTION'],
                    row['EMPLOYMENT TYPE']
                )

                # Execute INSERT statement
                with connection.cursor() as cursor:
                    cursor.execute(insert_query, values)

        # Commit changes to the database
        connection.commit()
        print("Data inserted successfully!")

    except Exception as e:
        print(f"Error inserting data into database: {e}")

    finally:
        # Close database connection
        if connection:
            connection.close()

# Path to your CSV file
csv_file_path = r'C:\Users\valen\Desktop\JOB-WEBSITE\output.csv'

# Call function to insert data from CSV into database
insert_data_from_csv(csv_file_path)
```
