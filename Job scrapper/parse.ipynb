```python
main_link= 'https://www.indeed.com/jobs?q=data+analyst+entry+level&l=Remote&vjk=eb304b6d79a1ae31'     # indeed main url 

# Initialize WebDriver
driver = webdriver.Chrome(chrome_driver_path)
driver.get(main_link)
sleep(10)

indeed_cards = driver.find_elements_by_xpath(
    '//div[@class="job_seen_beacon"]')
print(indeed_cards)

my_card_link_list=[]
indeed_card_list=[]
sleep(7)
heads = "https://www.indeed.com"

sleep(7)
for WebElement in indeed_cards:
    sleep(5)
    elementHTML = WebElement.get_attribute('outerHTML')
    sleep(5)
    elementsoup = BeautifulSoup(elementHTML, 'html5lib')
    sleep(6)
    card_link= elementsoup.find('a',href = True)
    card_join = (card_link['href'])
    my_card_link = heads + card_join
    my_card_link_list.append(my_card_link)
print(my_card_link_list)
    
```

```python
position_list = []
company_name_list = []
location_list = []
salary_list = []
job_link_list = []
benefit_list = []
description_list  = []
employment_type_list= []


for lines in my_card_link_list:
    driver.get(lines)
    sleep(2)
    html_content = driver.page_source
    sharksoup = BeautifulSoup(html_content, 'html.parser')
    
    # Position
    linkz = sharksoup.find('div', {'class': 'jobsearch-InfoHeaderContainer jobsearch-DesktopStickyContainer css-zt53js eu4oa1w0'})
    position = linkz.find('h1').text.strip() if linkz and linkz.find('h1') else "NULL"
    position_list.append(position)

    # Company Name
    company_name = linkz.find('span', {'class': "css-1saizt3 e1wnkr790"}).text.strip() if linkz and linkz.find('span', {'class': "css-1saizt3 e1wnkr790"}) else "NULL"
    company_name_list.append(company_name)

    # Location
    location = linkz.find('div', {'class': "css-1ikmi61 eu4oa1w0"}).text.strip() if linkz and linkz.find('div', {'class': "css-1ikmi61 eu4oa1w0"}) else "NULL"
    location_list.append(location)

    # Salary
    salary = linkz.find('span', {'class': "css-19j1a75 eu4oa1w0"})
    salary_text = salary.text.strip() if salary else "NULL"
    salary_list.append(salary_text)

    # Job Link
    # Apply_link =sharksoup.find('div', {'class': "jobsearch-ViewJobButtons-container css-15enpd4 eu4oa1w0 jobsearch-DesktopStickyContainer-buttonscontainer"})
    # button = Apply_link.find('button', class_='css-1oxck4n e8ju0x51')
    # if button:
    #     job_link = button['href']
    #     job_link_list.append(job_link)

    Apply_link = sharksoup.find('div', {'class': "jobsearch-ViewJobButtons-container css-15enpd4 eu4oa1w0 jobsearch-DesktopStickyContainer-buttonscontainer"})
    button = Apply_link.find('button', class_='css-1oxck4n e8ju0x51')
    job_link = button['href']
    job_link_list.append(job_link)    

    #Employment Type
    employment = sharksoup.find('div', {'class': "js-match-insights-provider-xu0md5 eu4oa1w0"})
    employ = employment.find('div', {'class': "js-match-insights-provider-tvvxwd ecydgvn1"}).text.strip() if employment and employment.find('div', {'class': "js-match-insights-provider-tvvxwd ecydgvn1"}) else "NULL"
    employment_type_list.append(employ)



    # Benefits
    benefit = sharksoup.find('div', {'id': "benefits", 'data-testid': "benefits-test", 'class': "css-jaxxzt eu4oa1w0"})
    benefits_list = []
    if benefit:
        ul_element = benefit.find('ul', {'class': 'css-8tnble eu4oa1w0'})
        if ul_element:
                benefits_list = [li_element.text.strip() for li_element in ul_element.find_all('li')]
    benefit_list.append(benefits_list if benefits_list else "No benefits found.")

    # Description
    description = sharksoup.find('div', {'id': "jobDescriptionText", 'class': "jobsearch-jobDescriptionText jobsearch-JobComponent-description css-10og78z eu4oa1w0"}).text.strip()
    description_list.append(description if description else "NULL")

print(position_list)
print(company_name_list) 
print(location_list) 
print(salary_list) 
print(job_link_list) 
print(benefit_list) 
print(description_list) 
print(employment_type_list)

```

```python
TABLE = {'POSITION':position_list,'COMPANY NAME': company_name_list,"LOCATION":location_list,'SALARY':salary_list,"JOB LINK":job_link_list,"BENEFITS":benefit_list,'DESCRIPTION':description_list,'EMPLOYMENT TYPE':employment_type_list}
daf = pd.DataFrame(TABLE)
daf
```
